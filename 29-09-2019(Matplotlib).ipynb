## matplotlib

### It is the lib function for ploting graphs based on your data 

- plot
- pie
- bar
- distplot
- boxplot

```python
import matplotlib.pyplot as mp

mp.plot([4,9,2],[1,5,7])

mp.show()
```

```python
k=list(range(1,10))

p=[i**2 for i in range(len(k))]

mp.plot(k,p,'r*')
```

```python
import numpy as np

x=list(range(0,30))
y=np.sin(x)

mp.plot(x,y,'g--')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)

mp.plot(x,y,"yd")
mp.title("Sine graph")
mp.xlabel("X axis")
mp.ylabel("y axis")
mp.grid()
```

```python
# pie chars

p=[14,56,63,32]

mp.pie(p)
```

```python
# Bar graph

p=list(range(6))

y=[74,18,52,9,63,35]

mp.bar(p,y)
```

```python
p=list(range(6))

y=[74,18,52,9,63,35]
z=[63,25,32,0,48,26]

mp.bar(p,y)
mp.bar(p,z)
mp.legend("Plot1","plot2")
```
